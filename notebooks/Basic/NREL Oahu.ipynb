```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
import math

import os
import sys
import glob
```

```python
input_path = '/Users/cseveriano/Documents/NREL'
output_path = '/Users/cseveriano/spatio-temporal-forecasting/data/processed/NREL/Oahu'
```

```python
all_dirs = glob.iglob(input_path+"/*", recursive=False)

for dir_ in all_dirs:
    print("Reading directory: ",dir_)
    if os.path.isdir(dir_):

        frame = pd.DataFrame()

        all_files = glob.iglob(dir_ + "/*.txt", recursive=False)
    
        for file_ in all_files:
            print("Reading: ",file_)
            df = pd.read_csv(file_, header=None, sep=",")
            df['datetime'] = df.apply(lambda row: pd.to_datetime(str(int(row[1]))+"/" + str(int(row[2])) + " " + str(int(row[3])) + " " + str(int(row[0])), format='%Y/%j %H%M %S'), axis=1)
            df.set_index('datetime', inplace=True)
            df = df.resample('15min').mean()        
            frame = frame.append(df)
        
        del df
        
        frame.drop(columns=[0,1,2,3], inplace=True)
        frame.columns = ['DHHL_3','DHHL_4', 'DHHL_5', 'DHHL_10', 'DHHL_11', 'DHHL_9', 'DHHL_2', 'DHHL_1', 'DHHL_1_Tilt', 'AP_6', 'AP_6_Tilt', 'AP_1', 'AP_3', 'AP_5', 'AP_4', 'AP_7', 'DHHL_6', 'DHHL_7', 'DHHL_8']
        
        filename = os.path.basename(dir_)
        
        print("Writing: ",output_path + "/" + filename)
        frame.to_csv(output_path + "/" + filename + ".txt", sep = ',')
        del frame
        
```
